This code shows how Principal Component Analysis (PCA) is used for dimensionality reduction

numpy: For numerical operations and generating random data.
PCA from sklearn.decomposition: Implements Principal Component Analysis.
matplotlib.pyplot: For visualizing data.

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

np.random.rand(100, 3): Generates a random dataset of shape (100, 3) where each sample has 3 features.
np.random.seed(42): Ensures the random data is the same each time the script runs (reproducibility).

```python
np.random.seed(42)
data = np.random.rand(100, 3)
```

PCA(n_components=2):

    Initializes PCA to reduce the data from 3 features to 2 features.
    Finds directions (principal components) that capture the most variance in the data.

fit_transform(data):

    Fits the PCA model to the data.
    Transforms the data into the new 2D space.

```python
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(data)
```

Shows how PCA transforms the high-dimensional data into a lower-dimensional representation while preserving the most important information (variance).

```python
plt.scatter(data[:, 0], data[:, 1], label='Original Data')
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], label='Reduced Data', marker='x')
plt.legend()
plt.title("Dimensionality Reduction with PCA")
plt.show()
```
